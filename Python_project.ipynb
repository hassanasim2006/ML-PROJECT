# CORRELATION MATRIX OF DATASET

```python
# creating a correlation matrix between all of the numerical data used in the given data
import pandas as pd
import numpy as np
# Assuming 'df' is your DataFrame (loaded from the Excel file)
# Select only numerical columns for correlation
numerical_df = df.select_dtypes(include=np.number)
# Calculate the correlation matrix
correlation_matrix = numerical_df.corr()
correlation_matrix
```

# FINDING HIGHEST AND LOWEST RELATION

```python

# Assuming 'correlation_matrix' is already calculated as in provided code
# Remove diagonal part (correlation of a variable with itself is always 1)
correlation_matrix = correlation_matrix.mask(np.equal(*np.indices(correlation_matrix.shape)))
# Find the highest correlation
max_value = np.nanmax(correlation_matrix)
row_max, col_max = correlation_matrix[correlation_matrix == max_value].stack().idxmax()
# Find the lowest correlation
min_value = np.nanmin(correlation_matrix)
row_min, col_min = correlation_matrix[correlation_matrix == min_value].stack().idxmin()
print(f"Highest correlation: {max_value} between '{row_max}' and '{col_max}'")
print(f"Lowest correlation: {min_value} between '{row_min}' and '{col_min}'")
```

# SOLUTION OF NULL ENTRIES

```python
# replace all entries in engine size (L) which are less than 1 with null entry

# Assuming 'df' is your DataFrame
df.loc[df['Engine Size(L)'] < 1, 'Engine Size(L)'] = None
```

# FOR NUMERICAL DATA

```python
# if none entry is found in engine size column then replace it with the average value of taht column

# Calculate the average of the 'Engine Size(L)' column, excluding NaN values
average_engine_size = df['Engine Size(L)'].mean()

# Fill NaN values in the 'Engine Size(L)' column with the calculated average
df['Engine Size(L)'].fillna(average_engine_size, inplace=True)
```

# FOR STRINGS

```python
# if a null entry is found in make column the fill it with the most repeated entry in the whole column

# Find the most frequent value in the 'Make' column
most_frequent_make = df['Make'].mode()[0]

# Fill NaN values in the 'Make' column with the most frequent value
df['Make'].fillna(most_frequent_make, inplace=True)
```
